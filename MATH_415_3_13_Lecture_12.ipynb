```python
#Using odeint
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt


def f(x,y):
  return .2*x*y
h = .1
start = 1
stop = 1.5
n = int((stop - start)/h + 1)
x = np.linspace(start, stop, n)
y0 = 1
soln = odeint(f, y0, x)
print(soln)
plt.plot(x,soln)
plt.grid()
plt.title("Soln to dy/dx = .2xy, y(1)=1")
plt.show()
```
