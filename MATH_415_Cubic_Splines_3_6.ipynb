```python
import numpy as np
import matplotlib.pyplot as plt

# f(x) = {-x^2, x <= 0
#        { x+1, x > 0

def f(x):
  #piecewise defined function
  if x <= 0:
    return -x**2
  else:
    return x + 1

x = np.linspace(-2, 5, 1000)
n = len(x)
func = np.zeros(n)
for i in range(n):
  func[i] = f(x[i])
plt.plot(x,func)
plt.grid()
plt.show()
```

```python
from sympy import *

xx = Symbol("xx")
sym_f = Piecewise(
    (-xx**2, xx <= 0),
    (xx+1, xx > 0),
    (0, True)
  )
plot(sym_f, (xx, -2, 5))
```
