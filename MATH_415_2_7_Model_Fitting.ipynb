```r
#Example 1:
x <- c(1, 2.3, 3.7, 4.2, 6.1, 7)
y <- c(3.6, 3, 3.2, 5.1, 5.3, 6.8)
plot(x,y, pch = 16, col = 'blue')
grid()
title("Data with Linear Model")
#Apply Linear Least Squares Fit to data
model <- lsfit(x,y)
attributes(model) #What info is available from the solver
a <- model$coefficients[2]
b <- model$coefficients[1]
abline(model, col = 'purple') #plot the model on top of the data
#predict at 2.25
prediction <- a*2.25+b
#print(prediction)
# residuals = actual errors
model$residuals
#Plot residuals
plot(model$residuals, pch = 16, main = "Residuals")
grid()
#identify the absolute deviation
print(max(abs(model$residuals)))
```
