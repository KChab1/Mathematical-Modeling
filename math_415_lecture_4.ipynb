```python
import numpy as np
import matplotlib.pyplot as plt

n = 100 #number of iterations
a0 = 1000 #initial value
plt.scatter(0,a0)
a = np.zeros(n)#initialize array of zeros
a[0] = a0
for i in range(n-1): #iterate from 0 to n-1
  a[i+1] = -1.01*a[i]
  plt.scatter(i+1, a[i+1])
plt.grid()
plt.show()
```

```python
#|r|<1
a0 = .3
n = 25
plt.scatter(0,a0)
a = np.zeros(n)
for i in range(n-1):
  a[i+1] = .5*a[i] + .1
  plt.scatter(i+1,a[i+1])
plt.grid()
plt.show()
```

```python
#|r|>1
a0 = 100000
n = 25
plt.scatter(0,a0)
a = np.zeros(n)
for i in range(n-1):
  a[i+1] = 1.01*a[i] - 1000
  plt.scatter(i+1,a[i+1])
plt.grid()
plt.show()
```
