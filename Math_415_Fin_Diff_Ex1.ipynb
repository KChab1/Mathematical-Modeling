```python
import numpy as np
import matplotlib.pyplot as plt
#This code solves a second order DE of the form
#y''(t)+P(t)y'(t)+Q(t)y(t)=f(t)
a = 1
b = 2
n = 10
h = (b-a)/n
P = 3
Q = 2
t = np.linspace(1,b,n+1)
A = np.zeros([n-1,n-1])
b = np.zeros(n-1)
#initialize solution vector y
y = np.zeros(n+1)
#add in boundary points
y[0] = 1
y[-1] = 6
for i in range(n-1):
  A[i,i] = 2*h**2 - 2
for i in range(n-2):
  A[i,i+1] = 1 + (3*h)/2
  A[i+1, i] = 1 - (3*h)/2
#create vector b
for i in range(n-1):
  b[i] = 4*h**2*t[i+1]**2
#add boundary terms
b[0] = b[0] - (1-(3*h)/2)*y[0]
b[-1] = b[-1] - (1+(3*h)/2)*y[-1]

y2 = np.matmul(A,b)
y[1:-1] = y2
print(y)
print(y2)
plt.plot(t,y)
plt.grid()
plt.xlabel('t')
plt.ylabel('x')
plt.show()
```
