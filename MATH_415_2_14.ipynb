```r
#Ex 1 fit using nonlinear least squares
x <- seq(.5, 2.5, by = .5)
y <- c(.7, 3.4, 7.2, 12.4, 20.1)
plot(x,y, pch = 16, col = "purple")
grid()

#.5**2*.7+1*3.4+1.5**2*7.2+4*12.4+2.5**2*20.1 - numerator
#applying nonlinear least squares
yFit <- nls(y ~ a*x^2, start = list(a=0))
a <- coef(yFit)
#make a smoother curve
xx <- seq(0, 3, by = .1)
yy <- a*xx^2
lines(xx, yy)
#computing residuals, max abs deviation, SSR
print('Residuals: ')
residuals <- y - a*x^2
print(residuals)
print("Max Absolute Deviation ")
maxDev = max(abs(residuals))
print(maxDev)
print("Square of Residuals")
print(residuals^2)
SSR <- sum(residuals^2)
print(SSR)
```

```r
#Ex 2: fit to y = ax^n using transformed (linear) least squares
plot(x,y, pch = 16, col = "purple")
grid()
#plot(log(x), log(y), pch = 16, col = "purple")
model <- lsfit(log(x), log(y))
a_lin = model$coefficients[1]
a_lin = exp(a_lin)
n = model$coefficients[2]
lines(xx, a_lin*xx^n)
model$residuals
#computing residuals, max abs deviation, SSR
print('Residuals: ')
residuals_lin <- y - a_lin*x^n
print(residuals_lin)
print("Max Absolute Deviation ")
maxDev = max(abs(residuals_lin))
print(maxDev)
print("Square of Residuals")
print(residuals_lin^2)
print("SSR:")
SSR <- sum(residuals_lin^2)
print(SSR)
```
