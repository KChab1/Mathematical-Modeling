```python
#Kody Chabaud
#Mathematical Modeling
#Dr. Kuhn
#Homework 1:
#Starting with 500 as inital amount, pay 50 each month and multiply by 1.5% for
#interest. Compute total timeframe to pay off and graph data.

import numpy as np
import matplotlib.pyplot as plt

t0 = 500 #initial cost
n = np.arange(0,12,1) #Months in a year
rb = np.zeros(12) #Remaining balance to be derived

rb[0] = t0
for i in range(len(n)-1):
  rb[i + 1] = (rb[i] - 50) * 1.015
print(rb)

plt.grid()
plt.xlabel("Number of Months")
plt.ylabel("Balance Remaining")
plt.title("Time in Months to Pay Off Debt:")
plt.plot(n,rb, color = 'purple')
plt.show()

print("It will take", n[11], "months to pay off the total debt and interest")
```
