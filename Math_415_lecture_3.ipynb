```python
import numpy as np
import matplotlib.pyplot as plt


#n = np.array([0,1,2,3,4,5,6,7])
n = np.arange(0,8) #1-7, not inclusive
pn = np.array([9.6, 18.3, 29.0, 47.2, 71.1, 119.1, 174.6, 257.3])

plt.grid()
plt.xlabel("Time in Hours n")
plt.ylabel("Biomass pn")
plt.title("Yeast Population")
plt.scatter(n, pn)
plt.show()

#Computing Differences
change = np.diff(pn)
plt.grid()
plt.xlabel("Observed Biomass pn:")
plt.ylabel("Change in Biomass (Pn+1 - Pn):")
plt.title("Change in Biomass Vs. Biomass")
plt.scatter(pn[0:-1], change)
plt.show()

#(55.5-23.9)/(119.1-47.2) #Estimates K for K*Pn
```

```python
#EX 2:

#n = np.array([0,1,2,3,4,5,6,7])
n = np.arange(0,19,1) #1-7, not inclusive, scale of 1
pn = np.array([9.6, 18.3, 29.0, 47.2, 71.1, 119.1, 174.6, 257.3, 350.7, 441,
               513.3, 559.7, 594.8, 629.4, 640.8, 651.1, 655.9, 659.6, 661.8])
#Estimate k
k = .00083 #computed in class
#Plot y = kx
nonlinear_x = np.linspace(0, 20, 100)
pop0 = 9.6 #initial population
stop_val = 20 #stopping value
nn = np.arange(0, stop_val, 1)
pop = np.zeros(stop_val)
pop[0] = pop0
for i in range(stop_val -1): #Starts at zero, stops at 18
  pop[i+1] = pop[i] + k * (665-pop[i]) * pop[i] #fills in pop for 1 to 19

plt.plot(nn, pop, color = 'purple')

plt.grid()
plt.xlabel("Time in Hours n")
plt.ylabel("Biomass pn")
plt.title("Yeast Population")
plt.scatter(n, pn)
#plotting nonlinear model
plt.show()

#Computing Differences
change = np.diff(pn)
plt.grid()
plt.xlabel("Observed Biomass pn:")
plt.ylabel("Change in Biomass (Pn+1 - Pn):")
plt.title("Change in Biomass Vs. Biomass")
plt.scatter(pn[0:-1], change)
plt.show()

xvals = np.linspace(0, 120000, 10)
yvals = k* xvals
plt.grid()
plt.xlabel("(665*Pn)*Pn")
plt.ylabel("Change in Pn")
plt.title("(665*Pn)*Pn vs. Change in Pn")
plt.plot(xvals, yvals, color = 'purple')
plt.scatter((665-pn[0:-1])*pn[0:-1] , change)
plt.show()


1.01**120 * 1000
```
