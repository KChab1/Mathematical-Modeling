```python
#Kody Chabaud
#Mathematical Modeling Homework 5
#Dr. Kuhn
#Due 2-28-24

import numpy as np
import matplotlib.pyplot as plt
#x = int(input(""))
#Part 1
v = ([20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80])
d = ([42, 56, 73.5, 91.5, 116, 142.5, 173, 209.5, 248, 292.5, 343, 401, 464])
plt.scatter(v,d, color = 'cyan')
# plt.scatter(v,d, color = 'cyan')
# plt.plot(v, sureThing, marker = "*", color = 'purple')
# plt.grid()
# plt.show()

#Part 2

# def L1(x):
#   return (((x-25)*(x-30)*(x-35)*(x-40)*(x-45)*(x-50)*(x-55)*(x-60)*(x-65)*(x-70)*(x-75)*(x-80))/((20-25)*(20-30)*(20-35)*(20-40)*(20-45)*(20-50)*(20-55)*(20-60)*(20-65)*(20-70)*(20-75)*(20-80)))

# def L2(x):
#   return (((x-20)*(x-30)*(x-35)*(x-45)*(x-50)*(x-55)*(x-60)*(x-65)*(x-70)*(x-75)*(x-80))/((25-20)*(25-30)*(25-35)*(25-40)*(25-45)*(25-50)*(25-55)*(25-60)*(25-65)*(25-70)*(25-75)*(25-80)))

# ((25-20)*(25-30)*(25-35)*(25-40)*(25-45)*(25-50)*(25-55)*(25-60)*(25-65)*(25-70)*(25-75)*(25-80))

def divided_diff(v, d):
    n = len(d)
    coef = np.zeros([n, n])
    #first column is d values
    coef[:,0] = d
    for j in range(1,n):
        for i in range(n-j):
            coef[i][j] = (coef[i+1][j-1] - coef[i][j-1]) / (v[i+j]-v[i])

    return coef

divdif = divided_diff(v,d)[0, :]
#Take in divided difference and each v value from data set, multiply and add
def lagrange(divdif, x):
  return divdif[0] + divdif[1] * (x-20) + divdif[2] * (x-20) * (x-25) + divdif[3] *  (x-20) * (x-25) * (x-30) + divdif[4] *  (x-20) * (x-25) * (x-30) * (x-35) + divdif[5] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) + divdif[6] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) * (x-45) \
  + divdif[7] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) * (x-45) * (x-50) + divdif[8] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) * (x-45) * (x-50) * (x-55) + divdif[9] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) * (x-45) * (x-50) * (x-55) * (x-60) + divdif[10] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) * (x-45) * (x-50) * (x-55) * (x-60) * (x-65) \
  + divdif[11] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) * (x-45) * (x-50) * (x-55) * (x-60) * (x-65) * (x-70) + divdif[12] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) * (x-45) * (x-50) * (x-55) * (x-60) * (x-65) * (x-70) * (x-75)

#Test of lagrange from divided difference.
sureThing = ([lagrange(divdif, v[0]),
lagrange(divdif, v[1]),
lagrange(divdif, v[2]),
lagrange(divdif, v[3]),
lagrange(divdif, v[4]),
lagrange(divdif, v[5]),
lagrange(divdif, v[6]),
lagrange(divdif, v[7]),
lagrange(divdif, v[8]),
lagrange(divdif, v[9]),
lagrange(divdif, v[10]),
lagrange(divdif, v[11]),
lagrange(divdif, v[12])])

#plt.scatter(v,d, color = 'cyan')
plt.plot(v, sureThing, marker = "*", color = 'purple')
plt.grid()
plt.show()
print("Original d Values: ",d, "\n")
print("Divided Difference Linear Interpolation Values:", sureThing, "\n")

#Part 3:
# v <- c(20, 25, 30, 35, 40, 45, 50, 55, 60, 65, 70, 75, 80)
# d <- c(42, 56, 73.5, 91.5, 116, 142.5, 173, 209.5, 248, 292.5, 343, 401, 464)
# model <- lsfit(v,d)
# model$coefficients

#Using the above code, ran in R, we are able to find the lsfit of the data set and its coefficients. x ~ 6.89 in this example, so we will keep 7 coefficients. Not entirely sure what it means to get our coefficients from the lsfit, so taking a shot here. Notes don't show example of non smooth data.

def smooth(divdif, x):
   return divdif[0] + divdif[1] * (x-20) + divdif[2] * (x-20) * (x-25) + divdif[3] *  (x-20) * (x-25) * (x-30) + divdif[4] *  (x-20) * (x-25) * (x-30) * (x-35) + divdif[5] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) + divdif[6] *  (x-20) * (x-25) * (x-30) * (x-35) * (x-40) * (x-45)
roundTwo = ([smooth(divdif, v[0]),
smooth(divdif, v[1]),
smooth(divdif, v[2]),
smooth(divdif, v[3]),
smooth(divdif, v[4]),
smooth(divdif, v[5]),
smooth(divdif, v[6]),
smooth(divdif, v[7]),
smooth(divdif, v[8]),
smooth(divdif, v[9]),
smooth(divdif, v[10]),
smooth(divdif, v[11]),
smooth(divdif, v[12])])
plt.plot(v, d)
plt.scatter(v,roundTwo)
plt.grid()
plt.show()
print("Values of Smooth Set:", roundTwo, "\n")

#Part 4
print("lagrange guess of t = 79: ", lagrange(divdif,79), "\n")
print("smoothing guess of t = 79: ", smooth(divdif, 79), "\n")
```
