```python
#Kody Chabaud
#Homework 2
#Dr. Kuhn
#Due: 2/8/24
import numpy as np
import matplotlib.pyplot as plt
from tabulate import tabulate as tab

#4: 30 year mortgage
n = 360 # number of iterations
a0 = 100000 #starting value
a = np.zeros(n)
a[0] = a0
for i in range (n-1): #iterate from 0 to n-1
  a[i+1] = (a[i]-600) + (a[i]*.005)
  plt.scatter(i+1, a[i+1])
plt.xlabel("Months")
plt.ylabel("Remaining Balance")
plt.title("Remaining Balance of Loan Vs. Loan Age")
plt.grid()
plt.show()
print(a[359])
print("If they pay $600 per month consistently for 30 years, they will only have ", a[359], "Dollars","\n", "left on the loan by the 360th payment. a0 = 100000(no interest acrued yet) and a360 = $147.76" )
```

```python
#Question 1: Cipro
n = 30
a0 = 500
a = np.zeros(n)
a[0] = a0
#part a
for i in range(n-1):
  a[i+1] = .66*a[i]
  plt.scatter(i+1, a[i+1])
plt.grid()
plt.xlabel("Days Since First Dose Taken")
plt.ylabel("Remaining Cipro (mg)")
plt.title("Remaining Cipro From Initial 500 mg Dose")
plt.show()
print(a)

#part b
for i in range(n-1):
  a[i+1] = .66 * a[i] + 500
  plt.scatter([i+1], a[i+1])
print(a)
plt.grid()
plt.xlabel("Days Since First Dose Taken")
plt.ylabel("Remaining Cipro (mg)")
plt.title("Remaining Cipro From Taking New Tab Each Day")
plt.show()
print("\n","Based on the results, when only taking the first tab, the amount remaining in your body decreased in an exponential manner.", "\n" "After around 15 days, you have less than 1 mg of Cipro left in your system. ", "\n")
print("Based on the second part to this problem, whe needing to take a new pill each day, the amount remaining in your body grows exponentially." "\n", "Similarly, after 15 days, the amount consistently remains around 1470 mg in your system, so long as you take a new one each day.")
```
