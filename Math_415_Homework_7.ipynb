```python
#Kody Chabaud
#Mathematical Modeling
#Dr. Kuhn
#Due: 4/15/24

#Part 2:
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

#Create our system
def model(y, t):
    x, v = y
    dx = v
    dv = 0.06216 * x
    return [dx, dv]

#Our given initial conditions
y0 = [-2/3, 4/3]

t = np.linspace(0, 20, 100) #10 seconds, 100 points

#odeint solution
odeint_sol = odeint(model, y0, t)

#Euler's Method implementation
def euler_method(model, y0, t):
    y = np.zeros((len(t), len(y0)))
    y[0] = y0
    dt = t[1] - t[0]
    for i in range(1, len(t)):
        y[i] = y[i-1] + np.array(model(y[i-1], t[i-1])) * dt
    return y

euler_sol = euler_method(model, y0, t)

plt.plot(t, odeint_sol[:, 0], label='odeint solution')
plt.plot(t, euler_sol[:, 0], label="Euler's Method")
plt.title("odeint vs Euler's Method")
plt.xlabel('S (Time)')
plt.ylabel('Displacement in Feet')
plt.legend()
plt.grid()
plt.show()
```
