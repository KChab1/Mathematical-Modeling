```python
import numpy as np
import matplotlib.pyplot as plt

def P(x):
  L0 = -1/6*(x-5)*(x-6)*(x-7)
  L1 = 1/2*(x-4)*(x-6)*(x-7)
  L2 = -1/2*(x-4)*(x-6)*(x-7)
  L3 = 1/6*(x-4)*(x-5)*(x-6)
  return .2*L0+0*L1+1*L2+3*L3
x = np.array([4,5,6,7])
y = np.array([.2,0,1,3])
xmodel = np.linspace(4, 7, 100)
ymodel = P(xmodel)
plt.plot(xmodel, ymodel)
plt.grid()
plt.title("Lagrange Interpolating Polynomial")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
from scipy.interpolate import lagrange
coefs = lagrange(x,y) #lagrange coefficients. constant term is first(reverse order)
n = len(coefs)
model = coefs[3]*xmodel**3+coefs[2]*xmodel**2+coefs[1]*xmodel**1+coefs[0]
plt.scatter(x,y)
plt.plot(xmodel, model)
```
