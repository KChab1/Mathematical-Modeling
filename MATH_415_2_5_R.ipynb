slope C = (60.1-22.1)/(20.085-7.389) = 2.993 or about 3
So the model becomes y = 3*e^x
Predict at x = 5.1, y = 3e^(5.1) = 492

Another approach: Take the log of both sides:
y = C * e^x
ln(y) = ln(Ce^x)
ln(y) = ln(C) + ln(e^x)
ln(y) = ln(C) + x

```r
#After taking the log of y = Ce^x, we have the linear model ln(y) = ln(C)+x
plot(x, log(y), main = "Plot of (x, log(y))", xlim = c(0, 4), ylim = c(0, 6))
grid()
cat('x:',x)# Concatinate then print
print("")
cat('ln(y):', log(y))
```
