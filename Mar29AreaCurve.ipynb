```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,1,100)
y = np.exp(-x**3)
plt.plot(x,y)
plt.grid()
plt.show()
```

```python
#find the area
n=10000000
a=0
b=1
M=1
counter=0
for i in range(n):
    x=np.random.rand()   #pick a random point in range x:[0,1] and y:[0,M]
    y=np.random.rand()   #rand() is between 0 and 1 so no need to manipulate for this range
    if y<=np.exp(-x**3):
        counter += 1    #counter= counter+1
area=M*(b-a)*counter/n
print(area)
```

```python
#find the max M
x = np.linspace(-2,2,100)
y = -x**2+4
plt.plot(x,y)
plt.grid()
plt.show()
# M=4 works
```

```python
n=10000000
a=-2
b=2
M=4
counter=0
for i in range(n):
    x=a+(b-a)*np.random.rand()   #pick a random point in range x:[0,1] and y:[0,M]
    y=M*np.random.rand()   #rand() is between 0 and 1 so no need to manipulate for this range
    if y<=-x**2+4:
        counter += 1    #counter= counter+1
area=M*(b-a)*counter/n
print(area)
```

```python
#to plot the points on the range
x = np.linspace(-2,2,100)
y = -x**2+4
plt.plot(x,y, linewidth=2)
plt.grid()

n=1000
a=-2
b=2
M=4
counter=0
for i in range(n):
    xx=a+(b-a)*np.random.rand()   #pick a random point in range x:[0,1] and y:[0,M]
    yy=M*np.random.rand()   #rand() is between 0 and 1 so no need to manipulate for this range

    if yy<=-xx**2+4:
        counter += 1    #counter= counter+1
        plt.scatter(xx,yy, c='red')
    else:
        plt.scatter(xx,yy, c='green')

plt.show()
area=M*(b-a)*counter/n
print(area)
```
