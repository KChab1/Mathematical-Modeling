```r
#Kody Chabaud
#Math 415: Mathematical Modeling
#Dr. Kuhn
#Due: 2/22/24
#Homework 4
#Question 1, part a
x <- c(5,10, 20, 30, 40, 55, 60, 70, 80,90,100) # x = x * 10^-3
y <- c(0, 19, 57, 94, 134, 173, 216, 256, 297, 343, 390) # y = y * 10^-5
plot(x,y) #Base plot of x and y
grid()
title("Original Data Fitted with abline")
y_new = a*x+b
model <- lsfit(x,y)
a <- model$coefficients[2]
b <- model$coefficients[1]
residual1 <- y- a*x-b
maxresid1 <- max(abs(residual1))
cat("Maximum Residual of First Graph:", maxresid1)
abline(model, col = 'purple')
plot(x, y_new, col = 'purple') #New plot of x and y_new
grid()
title('Linear Least Squares Approximation:' )
cat("y_new values:",  y_new)

#Part b
yFit <- nls(y ~ a*x^2, start = list(a=0)) #nonlinear least squares approx. of form a*x^2
aa <-  coef(yFit)
yy <- aa*x^2 #assigning y value from equation
plot(x,yy)
grid()
title("Nonlinear Least Squares: Quadratic")
lines(x,yy, col = 'purple')
residual2 <- yy - aa*x^2
maxresid2 <- max(abs(residual2))
cat("Max Residual of Second Output:", maxresid2, "\n")
cat("New Y Values From Nonlinear Least Squares Approx:", yy)

#Part c
yNewFit <- nls(y ~ a*exp(b*x), start = list(a=0)) #nonlinear least squares approx of form a*e^(bx)
aaa <- coef(yNewFit)
yyy <- aaa*exp(b*x)
residual3 <- yyy - aaa*exp(b*x)
maxresid3 <- max(abs(residual3))
plot(x,yyy)
grid()
title("Nonlinear Least Squares: Exponential")
lines(x,yyy, col = 'blue')
cat("New Y Values From Nonlinear Approx of Exponential:", yyy, "\n")
cat("Max Residual of Final Graph: ", maxresid3)
```

Based on the data below (residuals), the 3rd graph, the one utilizing nonlinear quadratic methods, is the best graph to use as an accurate approximation of the data.
